## Excercise 3 - Constraints on parameters

### Unconstrained parameters

```python
model_gm1 = CmdStanModel(stan_file='code_4.stan')
out_gamma1 = model_gm1.sample(output_dir='samples',iter_sampling=6000,iter_warmup=1000, seed=4838282)
```

```python
out_gamma1.diagnose();
```

```python
N=500
xs = np.linspace(0,8,N)
pdfs = stats.gamma.pdf(xs, 1.25, scale = 1 / 1.25) 

plt.plot(xs, pdfs, linewidth=2)

## histogram of theta samples with 160 bins
theta = pd.Series(out_gamma1.stan_variable('theta'))
plt.hist(theta, bins=160, density=True, color="lightblue")

plt.gca().set_xlabel("theta")
plt.gca().set_ylabel("Probability Density Function")
plt.show()
```

### Constrained parameter

```python
model_gm2 = CmdStanModel(stan_file='code_5.stan')
out_gamma2 = model_gm2.sample(output_dir='samples',iter_sampling=6000,iter_warmup=1000, seed=4838282)
```

```python
out_gamma2.diagnose();
```

```python
N=500
xs = np.linspace(0,8,N)
pdfs = stats.gamma.pdf(xs, 1.25, scale = 1 / 1.25) 

plt.plot(xs, pdfs, linewidth=2)

## histogram of theta samples from the second model with 160 bins
theta = pd.Series(out_gamma2.stan_variable('theta'))
plt.hist(theta, bins=160, density=True, color="lightblue")

plt.gca().set_xlabel("theta")
plt.gca().set_ylabel("Probability Density Function")
plt.show()
```

#### Unconstrained parameters, in the context where some values do not make sense, result in high percentage of divergence during sampling (40% to 0.0083% for properly constrained parameter) 

## Excercise 4 -  Selection of parameters using equation solving



```python
model_tune = CmdStanModel(stan_file='code_6.stan')

F = len("Jakub")
L = len("Kosiski")
y0 = 1

data={'y_guess':[y0],
        'theta':[(F+L)/2]}
tunes = model_tune.sample(data=data, fixed_param=True, iter_sampling=1, iter_warmup=0, chains = 1, output_dir='sample')
```

```python
tunes.stan_variable('sigma')[0]
```

#### Stan provides means to solve algebraic equations. Here the standard deviation of half_normal distribution, such that with 99% probability samples from it will be less than 7 has been found. The result came out at 2.71757.

## Excercise 5 - different methods of defining models

```python
model_samp_st = CmdStanModel(stan_file='code_7.stan')
model_log_target = CmdStanModel(stan_file='code_8.stan')
model_log_target_ind = CmdStanModel(stan_file='code_9.stan')
data = {'N': F}
seed = 17011996
result_1 = model_samp_st.sample(data=data,seed=seed)
result_2 = model_log_target.sample(data=data,seed=seed)
result_3 = model_log_target_ind.sample(data=data,seed=seed)
```

```python
az.plot_density([result_1,result_2,result_3])
plt.show()
```

#### Values of theta have been similarly distributed for both ways of defining models. However, there is a difference between affecting individual distributions and using vectorized form to add log pdf to the joint log pdf  - in the first case, standard deviation seems to be smaller.
